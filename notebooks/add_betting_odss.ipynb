```python
import pandas as pd
import glob

# 1. Load your big merged dataset (2000–2020, no odds yet)
merged = pd.read_csv("../data/merged_file.csv")

# 2. Load all odds CSV files (2010–2020)
odds_files = glob.glob("../data/bet/*.csv")   # adjust path if needed
odds_dfs = []

for f in odds_files:
    df = pd.read_csv(f)
    
    # Keep only relevant columns (odds + keys to join)
    keep_cols = ["Date", "HomeTeam", "AwayTeam", "B365H", "B365D", "B365A"]
    df = df[[c for c in keep_cols if c in df.columns]]
    
    odds_dfs.append(df)

# Combine all odds files into one DataFrame
odds = pd.concat(odds_dfs, ignore_index=True)

# 3. Convert Date column to datetime in both datasets
merged["Date"] = pd.to_datetime(merged["Date"])
odds["Date"] = pd.to_datetime(odds["Date"])

# 4. Merge betting odds into main dataset
final = merged.merge(
    odds,
    on=["Date", "HomeTeam", "AwayTeam"], 
    how="left"   # keep all rows from merged, odds only if available
)

# 5. Save final dataset
final.to_csv("../data/merged_after_odds.csv", index=False)
```
